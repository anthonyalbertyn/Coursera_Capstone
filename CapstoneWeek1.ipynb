### Data extraction

The data will be extracted as follows:  

1. The research will filter the NYPD Complaint data by year (2019), level of crime (felony), description of crime (dangerous weapon), time of crime (between 7am to 9pm and between 2pm to 6pm) and crime successfull (true).

2. The latitude and longitude data for each of these felonies will then be used to do an API call to Fouresquare to find venues with category 'high school' within a radous of 500m from the crime.

3. A list of these high schools, including their venue id, name, city, latitude and longitude data will be compiled. None of these high schools will have no felonies near it because they were found by their distance from a felony.

4. The list of schools will then be cleaned and grouped by venue id, and a count column will be added - so we know how many crimes found this school as a venue

### Analysis
